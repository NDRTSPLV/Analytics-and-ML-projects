## Введение

В данной работе мы занимаемся построением алгоритма, который позволяет предсказывать цену автомобиля на основе некоторых его характеристик, таких как тип коробки, пробег и количество лошадиных сил. Для анализа мы использовали данные, полученные с открытого источника Kaggle (URL: https://www.kaggle.com/datasets/thedevastator/uncovering-factors-that-affect-used-car-prices).

В ходе работы мы выполнили следующие этапы: предобработку полученных данных, анализ данных, выбор модели для предсказания цены автомобиля и проверку выбранной модели на отложенной выборке.

Для лучшего понимания данных, приведем расшифровку названий столбцов:

- **dateCrawled**: дата выгрузки анкеты из базы;
- **name**: полное наименование автомобиля;
- **seller**: тип продавца (частник/дилер);
- **offerType**: тип предложения (продажа, ремонт и т.д.);
- **price**: цена автомобиля;
- **abtest**: тип теста (A/B);
- **vehicleType**: тип кузова автомобиля (внедорожник, седан и т.д.);
- **yearOfRegistration**: год регистрации автомобиля;
- **gearbox**: тип коробки переключения передач (ручная или автоматическая);
- **powerPS**: мощность автомобиля (в данной работе будем интерпретировать как лошадиные силы);
- **model**: модель автомобиля;
- **kilometer**: пробег автомобиля;
- **monthOfRegistration**: месяц регистрации автомобиля;
- **fuelType**: тип заправляемого топлива (дизель, бензин и т.д.);
- **brand**: марка автомобиля;
- **notRepairedDamage**: имеет ли автомобиль неустраненные повреждения;
- **dateCreated**: дата создания объявления;
- **nrOfPictures**: количество фотографий автомобиля в объявлении;
- **postalCode**: почтовый индекс хозяина автомобиля;
- **lastSeen**: дата последней активности пользователя.

В данном проекте для предсказания цены автомобиля мы применили как базовые модели (например, "Линейную регрессию", "Деревья решений" и "Случайный лес"), так и модели градиентного бустинга на основе библиотек LightGBM и CatBoost. Для поиска оптимальных параметров мы использовали методы, включающие в себя приемы кросс-валидации.

В качестве метрики мы выбрали метрику MAE, которая показывает точность наших прогнозов и отклонение предсказанных моделью значений от фактических значений.

### Обработка пропущенных значений

В ходе данного пункта мы выполним поиск пропущенных значений и их удаление или замену одним из возможных способов. Для начала приведем таблицу пропущенных значений в процентах от общего количества строк в таблице.

```python
column = 'Пропуски, %'

(
    pd.DataFrame(data.isna().mean() * 100, columns=[column])
    .style.set_precision(2)
    .background_gradient('coolwarm')
    
)
```

В данной работе мы решили не использовать модели заполнения пропусков, такие как KNNImputer, поскольку в данном случае это не является необходимостью. Кроме того, некоторые типы пропусков (случайные, неслучайные) не могут быть заполнены даже лучшими моделями. Поэтому мы решили самостоятельно заполнить пропуски, чтобы полностью понимать принцип заполнения и иметь ясное представление о том, каким образом были заполнены пропущенные значения в каждой строке.

Прежде всего, стоит отметить, что мы не сможем заполнить пропуски в столбце **notRepairedDamage** из-за ограниченной информации. Но и удаление строк с пропусками не является хорошим вариантом из-за большого количества пропущенных значений. Поэтому мы решаем заполнить пропуски в этом столбце значением 'unknown'.

Также поступим с пропусками в столбце "model". Хотя мы можем предположить модель автомобиля, исходя из его характеристик, у автопроизводителей часто есть похожие модельные ряды, поэтому заполнение пропусков значением 'unknown' будет более предпочтительным подходом.

Оставшиеся пропуски в столбцах **vehicleType**, **gearbox** и **fuelType** мы заполним медианой, исходя из значений в столбцах **brand** и **model**. Такой подход позволит нам заполнить пропуски на основе типовых характеристик автомобилей с определенными марками и моделями.

```python
# Заполнение пропусков в столбцах notRepairedDamage и model
data[['notRepairedDamage', 'model']] = \
data[['notRepairedDamage', 'model']].fillna('unknown')
```

```python
# Заполнение пропсуков в столбцах vehicleType, gearbox, fuelType
data['vehicleType'] = data.groupby(['brand', 'model']).vehicleType.apply(
lambda x: x.fillna(x.mode()[0]))

data['gearbox'] = data.groupby(['brand', 'model']).gearbox.apply(
lambda x: x.fillna(x.mode()[0]))

data['fuelType'] = data.groupby(['brand', 'model']).fuelType.apply(
lambda x: x.fillna(x.mode()[0]))
```

```python
# Просмотр количества пропусков после обработки
(
    pd.DataFrame(data.isna().mean() * 100, columns=[column])
    .style.set_precision(2)
    .background_gradient('coolwarm')
    
)
```

В результате обработки пропущенных значений нам удалось полностью исключить пропуски в нашем сете без удаления строк, что является большим плюсом. Так как сохраняя данные мы даем нашей модели больше примеров на которых она сможет обучиться, а следовательно ее предсказания станут качественее.

## Обучение моделей. Выбор лучшей модели

### Подготовка к обучению

Поскольку обучение моделей мы будем проводить с помощью RandomizedSearchCV, который включает в себя метод кросс - валидации то наши данные мы будем делить на тренировочную и отложенную выборку.

Целевым признаком в нашем сете является столбец Price, все остальные столбцы - признаки. Так же поскольку мы будем обучать модели градиентного бустинга, такие как LightGBM и CatBoost то сразу определим категориональные переменные и продублируем наши выборки для обучения данных моделей, приведя столбцы с типом object к типу category.

```python
# Полуим признаки и целевой признаки
features = data.drop('price', axis=1)
target = data['price']

# Разделим данные на тренировочную и тестовоую выборку 
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE
)

# Сделаем копию выборок для обучения моделей градиентного бустинга
features_train_ct = features_train.copy()
features_test_ct = features_test.copy()
```

Далее поскольку в нашем сете имееются столбцы с типом данных ***object*** воспользуемся кодировкой данных столбцов с помощью метода OrdinalEncoder.

Так же воспользуемся методом маштабирования наших данных.

```python
# Определим какие столбцы необходимо отмаштабировать/закодировать
obj_columns = features_train.select_dtypes(include=np.object).columns.to_list()
dig_columns = features_train.drop(obj_columns, axis=1).columns.to_list()

print(f'Список столбцов типа object: {obj_columns}')
print(f'Список столбцов содержащих числовые значения; {dig_columns}')
```

Теперь отмаштабируем данные, которые мы будем передовать в модели градиентного бустинга, так как вероятно это сможет повысить метрику, а следовательно и количество более близких к реальным ответам предсказания во время обучения нашей модели.

```python
scaler = StandardScaler()
features_train_ct[dig_columns] = scaler.fit_transform(features_train_ct[dig_columns])
features_test_ct[dig_columns] = scaler.transform(features_test_ct[dig_columns])
```

```python
# Создадим объяект для передачи в pipeline
col_transform = make_column_transformer(
    (
        OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1),
        obj_columns
    ),
    (
        StandardScaler(),
        dig_columns
    ),
    remainder='passthrough'
)
```

### Обучение базовых моделей

```python
# Создадим функцию для обучения базовых моделей с помощью метода RandomizedSearchCV
def search(pipeline, param):
    """
    Функция для обучения базовых моделей методом RandomizedSearchCV

    :param pipeline: Пайплайн содержащий в себе необходимые этапы
    :type pipeline: sklearn.pipeline.Pipeline
    :param param: Параметры модели
    :type param: dict
    """
    global features_train, target_train

    random_search = RandomizedSearchCV(
        pipeline,
        param_distributions=param,
        scoring='neg_mean_absolute_error',
        random_state=RANDOM_STATE,
        n_jobs=-1
    )

    start = time.time()
    random_search.fit(features_train, target_train)
    end = time.time()
    
    fit_time = end - start
    best_score = random_search.best_score_ * -1
    best_param = random_search.best_params_
    
    model = named_steps = list(pipeline.named_steps)[1]
    print(f'Лучшее значение MAE у модели {model}: {best_score}\n\nЛучшие параметры у {model} {best_param}')
    print('\nВремя обучения модели {}: {} секунд'.format(model, fit_time))

    return random_search, fit_time, best_param, best_score
```

### Обучение моделей градиентного бустинга

Теперь обучим модели градиентного бустинга.

```python
# Скопируем и немного изменим нашу функцию для обучения базовых моделей
def search_boost(model, param):
    """
    Функция для обучения базовых моделей методом RandomizedSearchCV

    :param param: Параметры модели
    :type param: dict
    """
    global features_train_ct, target_train, obj_columns

    random_search = RandomizedSearchCV(
        model,
        param, 
        scoring='neg_mean_absolute_error',
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    
    features_train_ct = features_train_ct.copy()
    features_train_ct[obj_columns] = features_train_ct[obj_columns].astype('category')
    
    start = time.time()
    random_search.fit(features_train_ct, target_train)
    end = time.time()

    fit_time = end - start
    best_score = random_search.best_score_ * -1
    best_param = random_search.best_params_
    
    print(f'Лучшее значение MAE у модели {model}: {best_score}\n\nЛучшие параметры у {model} {best_param}')
    print('\nВремя обучения модели {}: {} секунд'.format(model, fit_time))
    

    return random_search, fit_time, best_param, best_score
```

#### Модель LGBMRegressor

```python
%%time
# Обучение модели LightGBM

model_lgbm = LGBMRegressor(random_state=RANDOM_STATE, verbose=20)

param_lgbm = {
        'n_estimators': range(100, 5000, 100),
        'learning_rate': np.logspace(-3, 0, 5),
        'max_depth': range(2, 30),
        'min_gain_to_split': range(2, 14, 2)
    }

rs_lgbm, time_lgbm, param_lgbm, score_lgbm = search_boost(model_lgbm, param_lgbm)
```

#### Модель CatBoostRegressor

```python
%%time
# Обучение модели CatBoost

model_cb = CatBoostRegressor(random_state=RANDOM_STATE, loss_function='MAE', verbose=20, cat_features=obj_columns)

param_cb = {
    'n_estimators': range(100, 400, 100),
    'depth': range(2, 14, 2)
}

rs_cb, time_cb, param_cb, score_cb = search_boost(model_cb, param_cb)
```
